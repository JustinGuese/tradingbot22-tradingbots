```python
from requests import get
from basebot22.basebot import BaseBot
from datetime import datetime, timedelta
```

```python
# put the above array into one line
allowedStocks = ["AAPL", "MSFT", "GOOG", "TSLA", "AMD", "AMZN", "DG", "KDP", "LLY", "NOC", "NVDA", "PGR", "TEAM", "UNH", "WM", "URTH", "IWDA.AS", "EEM", "XAIX.DE", "BTEC.L", "L0CK.DE", "2B76.DE", "W1TA.DE", "RENW.DE", "BNXG.DE", "BTC-USD", "ETH-USD", "AVAX-USD", "GLD", "TMF", "FAS", "TQQQ", "QQQ", "UUP", "META", "PYPL", "ADBE", "UPRO", "BSV", "SQQQ", "NTSX", "DBMF", "VDE", "VNQ", "VHT", "VFH", "VOX", "VPU", "VAW", "VGT", "VIS", "VDC", "VCR", "VLUE", "FNDX", "VTV", "RWL", "DBA", "SHV", "DBB", "DBO", "URA", "WOOD", "DBE"]
```

```python
datas = dict()
bot = BaseBot("testbot")
for ticker in allowedStocks:
    datas[ticker] = bot.getData(ticker, start_date = datetime.now() - timedelta(days=int(365*3)))

# create one array only contianign closing prices of all of them
prices = dict()
for ticker in allowedStocks:
    prices[ticker] = datas[ticker]["adj_close"]
```

```python
import pandas as pd

# merge prices dfs into one
big = pd.DataFrame(prices)
big = big.fillna(method='ffill')
big = big.fillna(method='bfill')
big.head()
```

```python
big.corr()
```

```python
correlationpairs = dict()
corrs = big.corr()
for ticker in allowedStocks:
    correlationpairs[ticker] = {
        "highest" : {"ticker" : corrs[ticker].sort_values(ascending=False).index[1], "value" : corrs[ticker].sort_values(ascending=False)[1]},
        "lowest" : {"ticker" : corrs[ticker].sort_values(ascending=True).index[1], "value" : corrs[ticker].sort_values(ascending=True)[1]}
    }
```

```python
from pprint import pprint

pprint(correlationpairs)
```

## some observations

2B76.DE < robotics ETF, highest correlation with "information tech ETF", lowest with "bonds"



```python
from pypfopt import EfficientFrontier, EfficientCDaR
from pypfopt import risk_models
from pypfopt import expected_returns
from pypfopt import objective_functions
# Calculate expected returns and sample covariance
mu = expected_returns.mean_historical_return(big)
S = risk_models.sample_cov(big)
```

```python
# Optimize for maximal Sharpe ratio
ef = EfficientFrontier(mu, S)
ef.add_objective(objective_functions.L2_reg, gamma=1)
raw_weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
# ef.save_weights_to_file("weights.csv")  # saves to file
print(cleaned_weights)
ef.portfolio_performance(verbose=True)
```

```python
cleaned_weights = dict(cleaned_weights)
# remove all stocks with weight 0
cleaned_weights = {k: v for k, v in cleaned_weights.items() if v != 0}
# and sort them descending
cleaned_weights = dict(sorted(cleaned_weights.items(), key=lambda item: item[1], reverse=True))
cleaned_weights
```

```python
specificInvestment = dict()
for ticker, weight in cleaned_weights.items():
    howmuchusd = round(weight * 10000, 2) 
    if howmuchusd > 100:
        specificInvestment[ticker] = howmuchusd
    else:
        print("Not buying " + ticker + " because it is too small: " + str(howmuchusd))
pprint(specificInvestment)
```

```python
bot = BaseBot("manual-static-sharpeopt-l2")
for ticker, amount in specificInvestment.items():
    bot.buy(ticker, amount, amountInUSD=True)
```

```python
## try to group them by sharing highest lowest
highest = dict()
lowest = dict()

for ticker, data in correlationpairs.items():
    if data["highest"]["ticker"] not in highest:
        highest[data["highest"]["ticker"]] = []
    if data["lowest"]["ticker"] not in lowest:
        lowest[data["lowest"]["ticker"]] = []
    highest[data["highest"]["ticker"]] += [ticker]
    lowest[data["lowest"]["ticker"]] += [ticker]
    
# sort both by the length of the list (highest first)
highest = dict(sorted(highest.items(), key=lambda item: len(item[1]), reverse=True))
lowest = dict(sorted(lowest.items(), key=lambda item: len(item[1]), reverse=True))
    
print(highest)
```
