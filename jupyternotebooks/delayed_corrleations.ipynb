```python
TRADEABLE = [
  "AAPL",
  "MSFT",
  "GOOG",
  "TSLA",
  "AMD",
  "AMZN",
  "DG",
  "KDP",
  "LLY",
  "NOC",
  "NVDA",
  "PGR",
  "TEAM",
  "UNH",
  "WM",
  "CWEG.L",
  "IWDA.AS",
  "EEM",
  "BTC-USD",
  "ETH-USD",
  "AVAX-USD"
]
```

```python
import yfinance as yf
import pandas as pd
```

```python
df = yf.download(TRADEABLE, start="2019-01-01", end="2022-11-01", group_by="ticker", auto_adjust=True, threads=True)
df.head()
```

```python
df = df.bfill()
df = df.ffill()
df.head()
```

```python
df = df.swaplevel(axis = 1)
df = df["Close"]
df.head()
```

```python
for ticker in TRADEABLE:
    df[ticker + "_1week"] = df[ticker].shift(-7) # 1 week in the future
df.head()
```

```python
df.corr()["TSLA_1week"].sort_values(ascending=False)
```
